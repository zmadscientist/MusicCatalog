```python
import pandas as pd
import music_tag

def change_tags(file_path, new_tags):
    """
    Changes the tags of an audio file.

    Args:
        file_path (str): The path to the audio file.
        new_tags (dict): A dictionary of new tags, where keys are tag names and values are tag values.
    """
    try:
        f = music_tag.load_file(file_path)
        for tag_name, tag_value in new_tags.items():
            f[tag_name] = tag_value
        f.save()
        print(f"Successfully changed tags for {file_path}")
    except music_tag.exceptions.InvalidFileError:
        print(f"Error: Could not open or process {file_path}. Ensure it is a valid audio file.")
    except Exception as e:
        print(f"An error occurred: {e}")

path = "music/*.m*"

    
df = pd.read_excel('music/About.xlsx')

d_GenreAlbum= {}
for y in range(1940,2000):
    #print((y//10-190)*10)
    d_GenreAlbum[y] = f'{(y//10-190)*10}s'
for y in range(2000, 2026):
    #print((y//10)*10)
    d_GenreAlbum[y] = f'{(y//10)*10}s'
d_GenreAlbum

for index, row in df.iterrows():
    print(row['Filename'], row['Release Year'])
    file_path = f"music/{row['Filename']}"  # Replace with the actual file path
    new_tags = {
        'album': d_GenreAlbum[row['Release Year']],
        'year': row['Release Year'],
        'genre': d_GenreAlbum[row['Release Year']]
    }
    change_tags(file_path, new_tags)
```
